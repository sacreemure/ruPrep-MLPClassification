```python
import pandas as pd
```

```python
db = pd.read_csv('prep_phrases_db.csv')
db
```

```python
freq_labels = db['label'].value_counts()[:5].index.tolist()
freq_labels
```

```python
def form_preps_df(label):
    preps_df = db.loc[db['label'] == label]
    unique_preps = set(preps_df['prep'])
    preps_counts = preps_df.value_counts("prep")
    preps_freq_df = pd.DataFrame(preps_counts, columns=['frequency'])

    return preps_freq_df
```

```python
preps_df = form_preps_df('локатив')
plot = preps_df.plot(kind='bar', title='локатив')
preps_df.transpose()
```
